```python
torch.manual_seed(1234)
A = torch.rand(2,3).to('cuda')
torch.manual_seed(1234)
B = torch.rand(2,3).to('cuda')

A,B
```

```python
product = A @ B.T
product
```

```python
max_val = torch.max(product)
min_val = torch.min(product)

print(f'Max Value: {max_val}, Min Value: {min_val} ')
```
