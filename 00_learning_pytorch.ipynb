### Random Tensors

Why random tensors?

Random tensors are important beause, the way many neural network learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

`Start with random numbers -> Look at data -> update random numbers -> look at data -> update again and so on`

### Tensor Datatypes

**Note:** Tensor datatypes is one of the 3 big erros that you'll run into with pytorch and deep learning.

1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on right device.(if two tensors on differnet devices eg- one is on cpu and another on gpu(cuda), and they tries to interact with each other)

```python
float_32_tensor = torch.tensor((3.0,6.00,9.),
                               dtype = torch.int64, # what datatype is the tensor (eg. float32, float64, float16, int32...)
                               device = "cpu", #what device is the tensor on
                               requires_grad = False) # whether or not to track gradient with this tensor's operations

float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
```

```python
float_16_tensor
```

```python
float_16_tensor*float_32_tensor
```

#### Reshaping, stacking, sqeezing and unsqeezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* Stacking - combining multiple tensors on tops of each other (vstack) or side by side (hstack)
* Sqeeze - remove all `1` dimentions from a tensor
* unsqueezing - add a `1` dimention to a target tensor
* permute - return a view of the input with dimentions permuted

```python
x = torch.arange(1., 10., 0.5)
x,x.shape
```

```python
x_reshaped = x.reshape(3,2,3)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(2,3,3)
z, z.shape
```

### changing z changes x (because a view of a tensor shares the same memory as the original tensor)

```python
z[:,0] = 0
z,x
```

```python
# stacking tensors on top of each other
x = torch.arange(1,6)
x_stacked = torch.stack([x,x,x,x], dim = 0)
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x], dim = 1)
x_stacked
```

```python
a = torch.rand(3,3)
b = torch.rand(3,3)
c = torch.rand(3,3)

stacked = torch.stack([a,b,c], dim = 1)
stacked
```

# torch.unsqueeze() - adds a single dimention to a target tensor at a specific dimention

```python
print(f'previous target{y_squeezed }')
print(f'previous shape{y_squeezed.shape }')
```

## Moving tensors back to cpu

<p> if tensor is on GPU, can't transform it to numpy

```python
tensor_on_gpu.numpy()
```

<p> To fix the GPU with numpy issue, we can first set it to CPU

```python
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu
```
